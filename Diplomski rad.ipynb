```python
def landmarksOnScreen(image, points):
    #mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistics.FACEMESH_TESSELATION)
    #mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistics.POSE_CONNECTIONS)
    mp_drawing.draw_landmarks(image, points.left_hand_landmarks, mp_holistics.HAND_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(0,0,255), thickness=2, circle_radius=3),
                             mp_drawing.DrawingSpec(color=(255,0,0), thickness=2, circle_radius=2))
    mp_drawing.draw_landmarks(image, points.right_hand_landmarks, mp_holistics.HAND_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(255,0,0), thickness=2, circle_radius=3),
                             mp_drawing.DrawingSpec(color=(0,0,255), thickness=2, circle_radius=2))
```

```python
cap=cv2.VideoCapture(0)
with mp_holistics.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        _, image = cap.read()
        
        image= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        points = holistic.process(image)
        image.flags.writeable = True
        image= cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        landmarksOnScreen(image, points)
        
        cv2.imshow('OpenCv Feed', image)
        if cv2.waitKey(1) == ord('q') or cv2.getWindowProperty('OpenCv Feed', cv2.WND_PROP_VISIBLE) < 1:
            break;
    cap.release()
    cv2.destroyAllWindows()
```

# Prikupljanje podataka
- Kreiramo landmarkArray na gore predstavljeni nacin za ulazne podatke
- Funkcija koja dobijenu sliku pretvara u landmarkArray

```python
letters=['A', 'B', 'V', 'G', 'D', 'Dj', 'E', 'Zj',
         'Z', 'I', 'J', 'K', 'L', 'Lj', 'M', 'N', 
         'Nj', 'O', 'P', 'R', 'S', 'T', 'Cj', 'U',
         'F', 'H', 'C', 'Cz', 'Dz', 'Sj']
```

```python
for letter in letters:
    try:
        os.makedirs(os.path.join('Azbuka', letter))
    except:
        pass
```

```python
def buildLandmarkArray(image, holistic):
    image= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False
    points = holistic.process(image)
    image.flags.writeable = True
    image= cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    landmarkArrayLeft=np.zeros(21*3)
    if points.left_hand_landmarks:
        landmarkArrayLeft=(np.array(list(map(lambda lm: [lm.x, lm.y, lm.z], points.left_hand_landmarks.landmark)))).flatten()
        #print(points.left_hand_landmarks)
    landmarkArrayRight=np.zeros(21*3)
    if points.right_hand_landmarks:
        landmarkArrayRight=(np.array(list(map(lambda lm: [lm.x, lm.y, lm.z], points.right_hand_landmarks.landmark)))).flatten()
        #print(points.right_hand_landmarks)
    return points, np.concatenate([landmarkArrayRight, landmarkArrayLeft])
```

```python
cap = cv2.VideoCapture(0)

with mp_holistics.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    for letter in letters:
        for imageNum in range(51):
            _, image = cap.read()
            points, landmarkArray = buildLandmarkArray(image, holistic)

            landmarksOnScreen(image, points)
            cv2.putText(image, 'Images for ' + letter + ' num ' + str(imageNum), (30, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 1, cv2.LINE_AA)
            if imageNum == 0:
                cv2.putText(image, 'Get Ready', (200, 200),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 4, cv2.LINE_AA)
                cv2.imshow('Image with landmarks', image)
                cv2.waitKey(3000)
            else:
                cv2.imshow('Image with landmarks', image)
                p = os.path.join('Azbuka', letter, str(imageNum))
                np.save(p, landmarkArray)
                cv2.waitKey(500)

            if cv2.waitKey(1) == ord('q') or cv2.getWindowProperty('Image with landmarks', cv2.WND_PROP_VISIBLE) < 1:
                break

    cap.release()
    cv2.destroyAllWindows()
```

```python
cap.release()
cv2.destroyAllWindows()
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
result_map = {key: num for num, key in enumerate(letters)}
```

```python
sequences, results = [], []
for letter in letters:
    for imgNum in range(50):
        res = np.load(os.path.join('Azbuka', letter, "{}.npy".format(imgNum+1)))
        sequences.append(res) #ovime smo dobili listu nizova koji ce da predstavljaju ulaz
        results.append(result_map[letter]) #ovime smo dobili listu rezultata koji odgovaraju nizovima iz predhodne linije
```

```python
x = np.array(sequences, object)
y = to_categorical(results).astype(int)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.05)
```

# Model

```python
x_train.shape
```

```python
y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import TensorBoard
```

```python
log_dir = os.path.join('Logs')
tb_callback=TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
model.add(Dense(256, activation='selu', input_dim=126))
model.add(Dropout(0.5))
model.add(Dense(128, activation='selu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='selu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
x_train = x_train.astype(np.float32)
x_test = x_test.astype(np.float32)
```

```python
model.fit(x_train, y_train, epochs=300, callbacks=[tb_callback])
```

```python
model.save("signLanguage.h5")
```

# Glavni Deo Projekta

```python
current_letter = random.choice(letters)
gesture_detected = False
threshold = 0.999
sequence=""
cap = cv2.VideoCapture(0)
with mp_holistics.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        _, image = cap.read()
        points, landmarkArray = buildLandmarkArray(image, holistic)

        landmarksOnScreen(image, points)
        
        cv2.putText(image, 'Letter ' + current_letter, (30, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(image, sequence, (500, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 1, cv2.LINE_AA)
        
        res = model.predict(np.expand_dims(landmarkArray, axis=0))[0]
        
        if res[np.argmax(res)] > threshold:
            if(letters[np.argmax(res)]==current_letter):
                cv2.putText(image, 'Correct ' + current_letter, (200, 200),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 1, cv2.LINE_AA)
                cv2.imshow('Image with landmarks', image)
                cv2.waitKey(2000)
                current_letter = random.choice(letters)
                gesture_detected = False
            else:
                cv2.imshow('Image with landmarks', image)
        else:
            cv2.imshow('Image with landmarks', image)

        if cv2.waitKey(1) == ord('q') or cv2.getWindowProperty('Image with landmarks', cv2.WND_PROP_VISIBLE) < 1:
            break

    cap.release()
    cv2.destroyAllWindows()
```
